```python
#Importieren der benötigten Bibliotheken.  
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
```

```python
#Überblick über den importierten Datensatz. Die beiden Files wurden der einfachheitshalber in einem File zusammen gefügt. 
back_publications = pd.read_csv('DoblerYannik3.csv', on_bad_lines='skip', sep=';', encoding='utf-8-sig')
back_publications
```

```python
#Anpassen der Datensätze
back_publications['Tweet'] = back_publications['Tweet'].str.lower()
back_publications['titel'] = back_publications['titel'].str.lower()

df11 = back_publications.titel.str.split(expand=True).stack().value_counts().rename_axis('unique_values').reset_index(name='counts')
df22 = back_publications.Tweet.str.split(expand=True).stack().value_counts().rename_axis('unique_values').reset_index(name='counts')

dfm = pd.merge(df11, df22, on='unique_values', how='outer')

#Um defekte Daten (NaN) auszulassen wird 0 eingefügt
dfm = dfm.fillna(0)


#Vorbereiten der Zählung
dfm["grand_total"] = dfm["counts_x"] + dfm["counts_y"]
dfm.columns = ["word", "title", "tweet", 'grand_total']


#Wortrangliste beginnend beim Wort, das am meisten verwendet wurde
dfm = dfm.sort_values("grand_total",ascending=False)

#Wenn Liste neu sortiert, wird Index verschoben. Daher nochmals neu Definierung der Indexen.  
dfm = dfm.reset_index()


#Um die interessanten Buzz-Words herauszufinden, habe ich die folgende Liste manuel herausgesucht (Gesamthäufigkeit der Wörter von > 5):
#lst = ["mobile", "business", "unternehmen", "learning", "ioT", "forum", "digitalen", "Video", "Thema", "Digital", "Zukunft", "Studie", "Lernen", "digitale", "Zeit", "Social", "Enterprise", "E-Learning", "Kolumne", "Transformation", "Innovation", "Internet", "Technologien", "Web", "Digitalisierung", "Menschen", "Apps", "Media", 'Knowledge', 'Management', 'Google', 'Swiss', 'Online', 'Requirements', 'Wissensmanagement', 'Informations', 'Marketing', 'Web-Video', 'Praxis', 'Facebook', 'twitter', 'Crowdsourcing', 'Bildung', 'Transformation', 'Entwicklung', 'Executive', 'Universität' 'Student', 'Digitalisierung', 'Studierende', 'Studierenden', 'Keynote', 'Daten', 'Schweizer'] 
lst = ["mobile", "business", "unternehmen", "learning", "iot", "forum", "digitalen", "video", "thema", "digital", "zukunft", "studie", "lernen", "digitale", "zeit", "social", "enterprise", "e-learning", "kolumne", "transformation", "innovation", "internet", "technologien", "web", "digitalisierung", "menschen", "apps", "media", 'knowledge', 'management', 'google', 'swiss', 'online', 'requirements', 'wissensmanagement', 'informations', 'marketing', 'web-video', 'praxis', 'facebook', 'twitter', 'crowdsourcing', 'bildung', 'transformation', 'entwicklung', 'executive', 'universität' 'student', 'digitalisierung', 'studierende', 'studierenden', 'keynote', 'daten', 'schweizer'] 


#Sortieren der Liste
dfm = dfm[dfm['word'].isin(lst)]
dfm1 = dfm.sort_values("title",ascending=False)
dfm2 = dfm.sort_values("tweet",ascending=False)


```

```python
#Sortieren, beginnend mit den meisten Wörter verwendet in Publikationen

import matplotlib.pyplot as plt

fig, ax = plt.subplots()
plt.title ('Häufigste Worte - Tweets vs. Publikationen', fontsize=15, color="black")
plt.xlabel("Wörter", fontsize=14)
plt.ylabel('Frequenz', fontsize=14)
plt.xticks(rotation = 90)
fig.set_size_inches(20, 10)

w = 0.3
ax.bar(dfm1["word"], dfm1["title"],width=w, color='b', align='edge')
ax.bar(dfm1["word"], dfm1["tweet"],width=-w, color='g', align='edge')
ax.autoscale(tight=True)




```

```python
#Sortieren, beginnend mit den meisten Wörter verwendet in Tweets

import matplotlib.pyplot as plt

fig, ax = plt.subplots()
plt.title ('Häufigste Worte - Tweets vs. Publikationen', fontsize=15, color="black")
plt.xlabel("Wörter", fontsize=14)
plt.ylabel('Frequenz', fontsize=14)
plt.xticks(rotation = 90)
fig.set_size_inches(20, 10)

w = 0.3
ax.bar(dfm2["word"], dfm2["title"],width=w, color='b', align='edge')
ax.bar(dfm2["word"], dfm2["tweet"],width=-w, color='g', align='edge')
ax.autoscale(tight=True)

```
