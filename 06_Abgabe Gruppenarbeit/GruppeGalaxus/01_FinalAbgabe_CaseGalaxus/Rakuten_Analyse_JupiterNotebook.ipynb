### Datentypen

Im Datenset sind drei verschiedene Datentypen vorhanden. Erkennbar ist dies an der Linie "dtypes: datetime64(ns)(1), float64(3), object(3), " - In der Klammer ist ersichtlich, wie viele Variablen diesem Datentyp zugeordnet sind.

- float64 fassen Zahlen mit Gleitkomma Zahlenformat zusammen
- Object, auch String genannt, ist eine Kombination aus Zahlen oder/und Buchstaben (Zeichenketten) und muss immer in Gänsefüsschen ""/'' geschrieben werden
